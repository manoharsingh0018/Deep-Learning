```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sn
```

```python
df = pd.read_csv("sonar_dataset.csv", header = None)
df.sample(5)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df[60].value_counts()
```

```python
X = df.drop(60, axis = 'columns')
y = df[60]
X
```

```python
y = pd.get_dummies(y, drop_first = True)
y.sample(5) # R= 1 and M = 0
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train,y_test = train_test_split(X, y, test_size = 0.25, random_state = 1)
```

```python
X_train.shape, X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim = 60, activation = 'relu'),
    keras.layers.Dense(15, activation = 'relu'),
    keras.layers.Dense(1, activation = 'sigmoid')
])

model.compile(optimizer = 'adam',
              loss = 'binary_crossentropy',
             metrics = ['accuracy'])

model.fit(X_train, y_train, epochs = 100, batch_size = 8)
```

```python
#we achieved an accuracy of model of 100%
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

y_pred = np.round(y_pred)
print(y_pred[:10])
print(y_test[:10])
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test,y_pred))
```

```python
#dropout regulariszation
model = keras.Sequential([
    keras.layers.Dense(60, input_dim = 60, activation = 'relu'),
    keras.layers.Dropout(0.5),#0.5 represent dropout factor that is 0.5 means 50%neuron will drop
    keras.layers.Dense(30, activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation = 'sigmoid')
])

model.compile(optimizer = 'adam',
              loss = 'binary_crossentropy',
             metrics = ['accuracy'])

model.fit(X_train, y_train, epochs = 100, batch_size = 8)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred1 = model.predict(X_test).reshape(-1)
print(y_pred1[:10])

y_pred1 = np.round(y_pred1)
print(y_pred1[:10])
print(y_test[:10])
```

```python
print(classification_report(y_test,y_pred1))
```
